# **Happiness/Income Study**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

hapiness_data = pd.read_csv("hapiness_data.csv")

sorted_hapiness_data = hapiness_data.sort_values(by="avg_income", ascending=False)
```

### **Highest/Lowest Scoring Country On Live Satisfaction**

```python
sorted_satisfaction_data = sorted_hapiness_data.sort_values(by='avg_satisfaction', ascending=False)

most_happy_country = sorted_satisfaction_data.iloc[0]
least_happy_country = sorted_satisfaction_data.iloc[-1]

print(most_happy_country)
print('\n')
print(least_happy_country)
```

### **Lowest/Highest Scoring Country On Average Income**

```python

top_country = sorted_hapiness_data.iloc[-1]
bottom_country = sorted_hapiness_data.iloc[0]

print(top_country)
print(bottom_country)
```

```python
from sklearn.cluster import KMeans

sorted_hapiness_data = hapiness_data.sort_values(by="avg_income", ascending=False)

avg_satisfaction = sorted_hapiness_data["avg_satisfaction"]
avg_income = sorted_hapiness_data["avg_income"]

hapiness_income = np.column_stack((avg_satisfaction, avg_income))

kmeans_result = KMeans(n_clusters=3).fit(hapiness_income)

clusters = kmeans_result.cluster_centers_


plt.figure(figsize=(12, 6))
plt.scatter(avg_satisfaction, avg_income, label="Country", color="blue")
plt.scatter(clusters[:,0], clusters[:,1], s=500)
plt.xlabel("Average Satisfaction")
plt.ylabel("Average Income in USD")
plt.title("Income and Hapiness Correlation Graph")
plt.legend(loc="upper left")



""" plt.text(5.5, 5, "Cluster 1")
plt.text(4, 3, "Cluster 2") """

plt.show()
```
